# 使用 TensorFlow 实现 Nearest Neighbor Algorithm

```python
import numpy as np
import tensorflow as tf
```

```python
# 导入实验所需数据（MNIST）并读取

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
# 准备训练数据
X_tr, Y_tr = mnist.train.next_batch(5000) 

# 准备测试数据
X_te, Y_te = mnist.train.next_batch(200)

xtr = tf.placeholder(tf.float32, [None, 784])
xte = tf.placeholder(tf.float32, [784])

# 计算 L1 距离
distance = tf.reduce_sum(tf.abs(tf.add(xtr, tf.negative(xte))), reduction_indices=1)

pred = tf.arg_min(distance, 0)


init = tf.global_variables_initializer()
```

```python
accuracy = 0.

with tf.Session() as sess:
    sess.run(init)
    
    for i in range(len(X_te)):
        # 找到最近的邻居
        nn_index = sess.run(pred, feed_dict={xtr: X_tr, xte: X_te[i, :]})
        print("Test {} Prediction: {} True Class: {}".format(
            i, np.argmax(Y_tr[nn_index]), np.argmax(Y_te[i])))
        if np.argmax(Y_tr[nn_index]) == np.argmax(Y_te[i]):
            accuracy += 1./len(X_te)
            
    print("Done!")
    print("Accuracy: %f" % accuracy)
```
